```python
# 1- Criando duas variáveis e exibindo a soma

num1 = 10
num2 = 5
print("Soma:", num1 + num2)

# 2- Exibindo a multiplicação

multiplicacao = num1 * num2
print("Multiplicação:", multiplicacao)

# 3- Calculando a média de 3 notas

nota1 = 7.5
nota2 = 8.0
nota3 = 6.5
media = (nota1 + nota2 + nota3) / 3
print("Média:", media)

# 4- Verificando se num1 é positivo ou negativo

if num1 >= 0:
    print("num1 é positivo")
else:
    print("num1 é negativo")

# 5- Verificando se num2 é par ou ímpar

if num2 % 2 == 0:
    print("num2 é par")
else:
    print("num2 é ímpar")

# 6- Verificando a situação do estudante

if media >= 6.0:
    print("Aprovado")
elif 4.0 <= media < 6.0:
    print("Recuperação")
else:
    print("Reprovado")

# 7- Verificando a categoria etária

idade = 25
if idade < 12:
    print("Criança")
elif 12 <= idade < 18:
    print("Adolescente")
elif 18 <= idade < 60:
    print("Adulto")
else:
    print("Idoso")

# 8- Números de 1 a 10 usando while

i = 1
while i <= 10:
    print(i, end=" ")
    i += 1
print()

# 9- Números de 1 a 10 usando for

for i in range(1, 11):
    print(i, end=" ")
print()

# 10- Tabuada de num1

for i in range(1, 11):
    print(f"{num1} x {i} = {num1 * i}")

# 11- Números de 0 a 100 de 10 em 10 usando while

i = 0
while i <= 100:
    print(i, end=" ")
    i += 10
print()

# 12- Números pares de 0 a 20 usando for

for i in range(0, 21, 2):
    print(i, end=" ")
print()

# 13- Converção Celsius para Fahrenheit

def celsius_para_fahrenheit(c):
    return c * 1.8 + 32

temperatura_celsius = 25
print("Temperatura em Fahrenheit:", celsius_para_fahrenheit(temperatura_celsius))

# 14- Calculadora

def calculadora():
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    operacao = input("Digite a operação (+, -, *, /): ")
    
    if operacao == '+':
        print("Resultado:", num1 + num2)
    elif operacao == '-':
        print("Resultado:", num1 - num2)
    elif operacao == '*':
        print("Resultado:", num1 * num2)
    elif operacao == '/':
        if num2 != 0:
            print("Resultado:", num1 / num2)
        else:
            print("Erro: divisão por zero")
    else:
        print("Operação inválida")

calculadora()
```

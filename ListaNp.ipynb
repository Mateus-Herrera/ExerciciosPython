```python
import numpy as np

# Exercício 1:
arr1 = np.array([3, 5, 9, 16, 25], dtype=float)
print(arr1)

# Exercício 2:
arr2 = np.array([1.2, -3.6, 8.99999, 16.1], dtype=int)
print(arr2)

# Exercício 3:
matriz_zeros = np.zeros((3, 9))
print(matriz_zeros)

# Exercício 4:
matriz_uns = np.ones((6, 7), dtype=int)
print(matriz_uns)

# Exercício 5:
matriz_identidade = np.eye(4)
print(matriz_identidade)

# Exercício 6:
arr_random = np.random.rand(5)
print(arr_random)

# Exercício 7:
matriz_random = np.random.randint(0, 6, (3, 6))
print(matriz_random)

# Matriz para os exercícios 8 a 11
matriz = np.array([[4, 6, 8], [-4, 3, 2], [22, -15, 12]])

# Exercício 8:
print(matriz.max())

# Exercício 9:
print(matriz.min())

# Exercício 10:
print(matriz.sum())

# Exercício 11:
print(matriz.mean())

# Array exercícios 12 a 16
arr3 = np.array([13, -2, 6, 8, 23, 15, 3, 90, 4, 10, 56, 12])

# Exercício 12:
print(arr3[3])

# Exercício 13:
print(arr3[-3:])

# Exercício 14:
print(arr3[2:])

# Exercício 15:
print(arr3[1:6])

# Exercício 16:
print(arr3[4:8])

# Matriz exercícios 17 a 20
matriz2 = np.array([[1, 3, 16, -25, 14],
                    [2, 0, 10, 5, 17],
                    [19, 4, 8, 15, 13],
                    [7, 12, -5, 14, 0]])

# Exercício 17:
print(matriz2[2])

# Exercício 18:
print(matriz2[:, 3])

# Exercício 19:
print(matriz2[1, 2])

# Exercício 20:
print(matriz2[1:3, 1:4])

# Matriz último exercício
matriz3 = np.array([[2, -1, 15, 24],
                    [-3, 18, 4, 25],
                    [9, 10, -13, 7]])

# Matriz pares / ímpares
nova_matriz = np.where(matriz3 % 2 == 0, 2, 1)
print(nova_matriz)
```
